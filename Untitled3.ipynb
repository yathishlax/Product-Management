##replacing missing values

```
features_with_na = [features for features in share3.columns if share3[features].isnull().sum()>1]

for feature in features_with_na:
   
    meadian_valueDQ=share3['Deliverable Quantity'].median()
    share3['Deliverable Quantity']=share3['Deliverable Quantity'].fillna(meadian_valueDQ)
        
    median_valueDQT=share3['% Deli. Qty to Traded Qty'].median()
    share3['% Deli. Qty to Traded Qty']=share3['% Deli. Qty to Traded Qty'].fillna(median_valueDQT)
```

###visualizing for checking shewness

```
import scipy.stats as stats

def diagnostic_plots(share4, variable):
       
    plt.figure(figsize = (10,12))
    plt.subplot(1, 2, 1)
    share4[variable].hist()
    
    plt.subplot(1,2,2)
    stats.probplot(share4[variable], dist = "norm", plot = plt)
    plt.title(variable)
    plt.show()
```

##list comprehension for skewness

```
feat = [features for features in share4.columns]

for feature in feat:

    diagnostic_plots(share4,feature)
```

###there is a huge skewness and log transformation is required 

####separating positive and negative variable    

```
featpostive= [feature for feature in share4.columns if feature not in ['Date','Spread C-O', 'Nextdayclose']]

featnegative = share4[['Spread C-O']]
```

#####taking log transformation 

```
for feature in featpostive:
     
    if 0 in share4[feature].unique():
        share4['log'+feature]=share4[feature]
    else:
      share4['log'+feature] = np.log(share4[feature])

    
for feature in featnegative:
    
    if 0 in share4[feature].unique():
        share4['log'+feature]=share4[feature]
    else:
        share4['log'+feature] = (np.log (abs(share4[feature])))

   
featlog = [feature for feature in share4.columns if feature.startswith('log')]
```

#####visualizing for log transformation

```
for feature in featlog:
    diagnostic_plots(share4,feature)
```

###selecting required features

```
share5 = share4.iloc[:,12:26]

#concatinating
feat000 = [feature for feature in share5.columns if feature not in ['logIncresenext day']]
       
share5 = share5[feat000]
```

```
share5.isnull().sum() 
```

```
share6 = share5.copy()
```

### Identifying outliers

```
Q1 = share6.quantile(0.25)
Q3 = share6.quantile(0.75)

IQR = Q3 - Q1

share6_out = share6[~((share6 < (Q1 - 1.5 * IQR)) |(share6 > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```
share6 = share6_out.copy()
```

###Independent and dependent variable

```
X = share6.drop(['Incresenext day'], axis = 1)
Y = share6['Incresenext day']
```

###feature selection

```
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```
feature_sel_model = SelectFromModel(Lasso(alpha = 0.005, random_state=0))
feature_sel_model.fit( X,Y )

feature_sel_model.get_support()

selected_feat = X.columns[(feature_sel_model.get_support())]


X = X[selected_feat]
```

###train test split

```
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 1/3, random_state = 0)
```

###feature scaling

```
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

####importing libraries

```
import tensorflow.keras
from tensorflow.keras.models import Sequential#for hidden layers
from tensorflow.keras.layers import Dense, Activation, Embedding, Flatten,  BatchNormalization
from tensorflow.keras.layers import   ELU
from tensorflow.keras.layers import Dropout
```

```
tensorflow.keras.layers.LeakyReLU
tensorflow.keras.layers.ReLU
```

#initializing the ANN without hyper parameter tuning

```
classifier = Sequential()
```

####adding the input layer and first hidden layer

```
classifier.add(Dense(units = 1, kernel_initializer = 'he_uniform', activation = 'relu', input_dim = 3))

classifier.add(Dropout(0.3))
```

#####adding the second first hidden layer

```
classifier.add(Dense(units = 1, kernel_initializer = 'he_uniform', activation = 'relu'))

classifier.add(Dropout(0.4))
```

####adding the output layer

```
classifier.add(Dense(units = 1, kernel_initializer = 'glorot_uniform', activation = 'sigmoid'))
classifier.add(Dropout(0.2))
```

```
#compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

#fitting the ANN to training set
model_history = classifier.fit(X_train, Y_train, validation_split =0.33, batch_size = 10, epochs = 100)
```

```
classifier.summary()
```

```
#making predictions on evalating the initial model
Ypred = classifier.predict(X_test)
Ypred = (Ypred>0.5)
```

```
#making consfusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Ypred,Y_test )

#calclate 
from sklearn.metrics import accuracy_score
score = accuracy_score(Ypred, Y_test)
```

```
print(cm)
print(score)
```

#selecting hyper parameters 

```
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV

#with hyper parameter tuninig
#Selecting the neurons
def create_model(layers, activation):
    model = Sequential()
    for i, nodes in enumerate(layers):
        if i==0:
            model.add(Dense(nodes,input_dim=X_train.shape[1]))
            model.add(Activation(activation))
            model.add(Dropout(0.3))
        else:
            model.add(Dense(nodes))
            model.add(Activation(activation))
            model.add(Dropout(0.3))

            model.add(Dense(units = 1, kernel_initializer= 'glorot_uniform', activation = 'sigmoid')) # Note: no activation beyond this point
    
        model.compile(optimizer='adam', loss='binary_crossentropy',metrics=['accuracy'])
        return model
    
model = KerasClassifier(build_fn=create_model, verbose=0)


```

```
layers = [[20], [40, 20], [45, 30, 15]]
activations = ['sigmoid', 'relu']
param_grid = dict(layers=layers, activation=activations, batch_size = [128, 256], epochs=[30])
```

```
grid = GridSearchCV(estimator=model, param_grid=param_grid,cv=5)
```

```
grid_result = grid.fit(X_train, Y_train)
```

```
print([grid_result.best_score_,grid_result.best_params_])
```
